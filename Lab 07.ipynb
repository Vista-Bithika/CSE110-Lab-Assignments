```python
#Task 7
def show_palindrome(num):
    final = ""
    count = 0
    for i in range(1,num+1):
        i = str(i)
        final += i
        count += 1
    for i in range(1,num):
        count -= 1
        final += str(count)
    return final
print(show_palindrome(5))
```

```python
#Task 8
def show_palindrome(m):
  for i in range(1,m+1):
    print(i,end="   ")
  for i in range(m-1,0,-1):
    print(i,end="   ")

def show_palindromic_triangle(n):
  for i in range(1,n+1):
    print(" "*(n-i)*4,end="")
    show_palindrome(i)
    print()

show_palindromic_triangle(5)
```

```python
#Task 10
def make_square(a):
  start = a[0]
  end = a[1]

  a_dict = {}
  for i in range(start,end+1):
    a_dict[i] = i**2
  return a_dict

print(make_square((2,5)))
```

```python
#Task 11
def remove_duplicate(tuple_1):
    list_1 = []
    for i in tuple_1:
        if i not in list_1:
            list_1.append(i)
    list_1 = tuple(list_1)
    return list_1
print(remove_duplicate((1,1,1,2,3,4,5,6,6,6,6,4,0,0,0)))
print(remove_duplicate(("Hi", 1, 2, 3, 3, "Hi",'a', 'a', [1,2])))
```
