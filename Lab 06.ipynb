```python
#task 8
a=input().split(", ")
my_dict = {}
for i in a:
  for j in range(len(i)):
    if i[j] == ":":
      my_dict[i[0:j]] = int(i[j+1:])

l1 = list(my_dict.values())
c = 0
sum = 0
for i in l1:
  c += 1
  sum += i
avg = sum/c
print("Average is" ,int(avg))
```

```python
#task 8
a=input().split(",")
b={}

for i in a:
  l1=i.split(":")
  b[l1[0]]=int(l1[1])
sum=0 
count=0 
for i ,j in b.items():
  sum+=j
  count+=1
avg=sum/count
print(int(avg)) 
```
